```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 客户端数据，示例中只展示部分数据，你可以用全部数据替换
def fun(data):
    num_clients = len(data)
    num_classes = 10

    # 构造一个 num_clients x num_classes 的矩阵，每行代表一个客户端，每列代表一个类别
    matrix = np.zeros((num_clients, num_classes))
    for client_id, client_data in enumerate(data):
        for label, count in client_data:
            matrix[client_id, label] = count

    # 绘制热力图
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrix, annot=True, fmt="g", cmap="YlGnBu",
                xticklabels=[f"Class {i}" for i in range(num_classes)],
                yticklabels=[f"Client {i}" for i in range(num_clients)])
    plt.xlabel("classes")
    plt.ylabel("clients")
    plt.show()
```
